# Projeto: Predição de Insuficiência Cardíaca (Heart Failure Prediction Dataset)

## Objetivos:
O objetivo desse prejeto é realizar a construção de um modelo que faça a predição de insuficiência cardíaca de pacientes, utilizando modelos preditivos aplicados ao dataset 'Heart Failure Prediction Dataset' do Kaggle. O projeto envolve desde a análise exploratória dos dados até a validação do modelo. O dataset apresenta as seguinte colunas:

*   **Age:** Idade do paciente [anos].

*   **Sex:** Sexo do paciente [M: Masculino, F: Feminino].

*   **ChestPainType:** Tipo de dor no peito [TA: Angina Típica, ATA: Angina Atípica, NAP: Dor Não Anginal, ASY: Assintomático].

*   **RestingBP:** Pressão arterial em repouso [mm Hg].

*  **Cholesterol:** Colesterol sérico [mm/dl].

*   **FastingBS:** Glicemia em jejum [1: se Glicemia em jejum > 120 mg/dl, 0: caso contrário].

*   **RestingECG:** Resultados do eletrocardiograma em repouso [Normal: Normal, ST: anormalidade na onda ST-T (inversões da onda T e/ou elevação ou depressão do segmento ST > 0,05 mV), LVH: hipertrofia ventricular esquerda provável ou definitiva pelos critérios de 'Estes'].

*   **MaxHR:** Frequência cardíaca máxima alcançada [Valor numérico entre 60 e 202].

*   **ExerciseAngina:** Angina induzida por exercício [Y: Sim, N: Não].

*   **Oldpeak:** Oldpeak = ST [Valor numérico medido em depressão].

*   **ST_Slope:** Inclinação do segmento ST no pico do exercício [Up: ascendente, Flat: plano, Down: descendente].

*   **HeartDisease:** Target [1: doença cardíaca, 0: Normal].

# Criação da Rede Neural LSTM

#### Criação do modelo

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Input 
import tensorflow as tf

def construirModelo(time_steps=10, qtdNeuroniosLSTM=32, qtdNeuroniosDense=64, dropout=0.3, learningRate=0.003, loss='mean_squared_error', optimizer='Adam', **kwargs):
    n_features = X_data.shape[2] 
    model = Sequential([
        Input(shape=(time_steps, n_features)), 
        LSTM(qtdNeuroniosLSTM),   
        Dropout(dropout),
        Dense(qtdNeuroniosDense, activation='relu'),
        Dense(qtdNeuroniosDense, activation='relu'),
        Dropout(dropout),
        Dense(qtdNeuroniosDense, activation='relu'),
        Dense(qtdNeuroniosDense, activation='relu'),
        Dropout(dropout),
        Dense(16, activation='sigmoid'),
        Dense(1, activation='sigmoid')
    ])
        
    if optimizer == 'Adam':
        optimizerModel = tf.keras.optimizers.Adam(learning_rate=learningRate)
    else:
        optimizerModel = tf.keras.optimizers.RMSprop(learning_rate=learningRate)

    model.compile(
        loss=loss,
        optimizer=optimizerModel,
        metrics=['recall', 'accuracy']
    )

    return model
```

#### Cross Validation do Modelo

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score 
from scikeras.wrappers import KerasClassifier

crossModel = KerasClassifier(model=construirModelo, epochs=30, batch_size=15, verbose=0)

kf = KFold(n_splits=5, shuffle=True, random_state=42)

scores = cross_val_score(crossModel, X_data, Y_data, cv=kf, scoring='recall')

print(scores)
print("Média:", np.mean(scores))
```

# Tunagem de Hiper Parâmetro

#### Utilizando o Grid Search

```python
from sklearn.model_selection import GridSearchCV

# TIME_STEPS=10, qtdNeuroniosLSTM=32, qtdNeuroniosDense=64, dropout=0.3, learningRate=0.003, loss='mean_squared_error', optimizer='Adam'

param_grid = { # Lista / Dicionário de possíveis hiperparâmetros do modelo
    'model__time_steps': [10, 20],
    'model__qtdNeuroniosLSTM': [16, 32, 64, 128],
    'model__qtdNeuroniosDense': [8, 16, 32, 64, 128],
    'model__dropout': [0.2, 0.3],
    'model__learningRate': [0.001, 0.0001, 0.002],
    'model__loss': ['mean_squared_error', 'binary_crossentropy'],
    'model__optimizer': ['Adam', 'rmsprop'],
    'epochs': [10, 20],
    'batch_size': [15, 30]
}


grid = GridSearchCV(estimator=crossModel, param_grid=param_grid, cv=3, scoring='f1', n_jobs=-1)
grid.fit(X_train, Y_train)

Tunned_GB = grid.best_estimator_
y_predict = Tunned_GB.predict(X_teste)

print("Melhores parâmetros:", grid.best_params_)
print("Melhor F1-score na validação cruzada:", grid.best_score_)
```

## Criando Modelo

Com base nos hiperparâmetros um modelo será construído. A quantidade de iterações permitirá, devido a estocasticidade, encontrar um modelo que forneça uma boa performance.

Focamos na métrica F2-Score, que prioriza o recall quanto à precisão, mas sem cair em casos de alto recall com baixíssima precisão. 

O modelo obtido e alguns dados são armazenados para avaliações posteriores. 

```python
from sklearn.metrics import recall_score, accuracy_score, fbeta_score 


n_execucoes = 30
melhor_f2 = -1.0 
melhor_recall_associado = -1.0 
melhor_acc_associada = -1.0   
melhor_modelo = None
melhor_historico = None 


for i in range(n_execucoes):
    
    model_temp = construirModelo(
        time_steps=20, 
        qtdNeuroniosLSTM=16, 
        qtdNeuroniosDense=32, 
        dropout=0.3, 
        learningRate=0.002,
        loss='binary_crossentropy',
        optimizer='Adam'
    )
    
    history = model_temp.fit(
        X_train,
        Y_train, 
        epochs=20, 
        batch_size=30,
        validation_split=0.1,
        verbose=0
    )

    Y_predict_temp = model_temp.predict(X_teste, verbose=0)
    Y_predict_class_temp = (Y_predict_temp > 0.5).astype(int)
    
    #Calculo das metricas
    #F2, que irá priorizar o recall.
    f2_atual = fbeta_score(Y_teste, Y_predict_class_temp, beta=2) 
    
    recall_atual = recall_score(Y_teste, Y_predict_class_temp)
    acc_atual = accuracy_score(Y_teste, Y_predict_class_temp)
        
    if f2_atual > melhor_f2:
        melhor_f2 = f2_atual
        melhor_recall_associado = recall_atual
        melhor_acc_associada = acc_atual
        melhor_modelo = model_temp 
        melhor_historico = history 

if melhor_modelo is not None:
    Y_predict_final = melhor_modelo.predict(X_teste)
    Y_predict_class_final = (Y_predict_final > 0.5).astype(int)
```
